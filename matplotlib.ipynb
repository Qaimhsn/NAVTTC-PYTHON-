```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,2*np.pi,0.1)
a=np.sin(x)
b=np.cos(x)
c=a+b
plt.plot(a,marker='*',label="Sine",ms=5, mec='r', mfc='black', ls =':', linewidth="0.1",color='k')
plt.plot(b,marker="x",label="Cosine",ms=5, mec='r', mfc='green', ls =':', linewidth="0.1")
plt.plot(c,marker="o",label="Sum",ms=5, mec='r', mfc='black', ls ='--', linewidth="0.1")
plt.legend()
plt.show()
```
