```python
arr3=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
arr4=np.concatenate([arr3,arr2])
print(arr4)
```

```python
arr5=np.array([[1,2,3,4],
             [10,11,12,13]])
arr7=np.array([[5,6,7,8],
             [9,14,15,17]])
arr6=np.concatenate([arr5,arr7])
print(arr6)
i=np.stack([arr5,arr7],axis=1)
print(i)
ie=np.vstack([arr5,arr7])
print(ie)
ih=np.hstack([arr5,arr7])
print(ih)
```
